```python
# p(x) is a polynomial function 
# p(x) = 3 - 4x + 6x^2 + 2x^3

def poly_f(poly,x):
    # poly is a set of polynomial coefficients
    n = len(poly)       # number of terms
    px = poly[n-1]
    for i in range(n-1, 0, -1):
        px = (x * px) + poly[i-1]
        print('poly[%d] = %f, poly[%d] = %f, px = %f' %(i,poly[i],i-1,poly[i-1],px))
    #end for
    return px
#end function
```

```python
# using stopping criterion eps = 0.5 * 10^(2-n)
# where n is the number of significant figures
# e.g. n = 6, then eps = 0.5 * 10^(2-6) = 0.00005
x = 2
i_max = 1000000
sf = 4          # number of significant figures
eps = 0.5 * (10**(2-sf))
print('eps = %2.6f' %(eps))
ea = 100        # initial approximation error percentage
prev_fx = f(x)
print('prev_fx = %f' %(prev_fx))
for i in range(1,i_max):
    x = x + i
    fx = f(x)
    ea = np.abs((fx - prev_fx)/fx)*100
    print('x = %f f(x) = %f ea = %2.6f' %(x,fx,ea))
    if (ea < eps):
        print('ea < eps : (%2.6f < %2.6f)' %(ea, eps))
        break
    #end if
    prev_fx = fx
#end for
```

```python
def exp_taylor(x, n):
    # n is the number of terms
    # e^x = x^0/0! + x^1/1! + x^2/2! + ... + x^n/n!
    # sum = 0
    #         ##previous sum 
    # for i in range(0,n): ##HW2 speed this up to be polynomial recursive
    #     sum = sum + x**(i)/math.factorial(i)
        
    # #end for
    # return sum
    if n == 0:
        return 0
    elif n == 1:
        return 1 + x
    else:
        # factorial = math.factorial(n)
        # if(factorial == 0): return 0
        return exp_taylor(x, n-1) + x**n/math.factorial(n)

#end function
```

```python
def exp_taylor_recursive(x, n, term=1, factorial=1, k=1):
    if k == n:
        return term
    else:
        factorial *= k
        term = term * x / factorial
        return term + exp_taylor_recursive(x, n, term, factorial, k + 1)

# Example usage:
x = 1.8
n_max = 15 #lessthan 15 is not converge 
actual_exp_x = np.exp(x) # true value 
print('actual_exp_x(%f) = %3.6f' %(x,actual_exp_x))
for n in range(0,n_max):
    # n is the number of term
    taylor_exp_x = exp_taylor(x,n)
    print('n = %d : taylor approximation of exp_x(%f) = %3.6f' %(n,x,taylor_exp_x))
#end for
# result = exp_taylor_recursive(x, n)
# print(result)
```

```python
def f_taylor(x, n):
    # n is the number of terms
    # e^x = x^0/0! + x^1/1! + x^2/2! + ... + x^n/n!
    sum = 0
            ##previous sum 
    for i in range(0,n):
        sum = sum + x**(i)
    #end for
    return sum
#end function

```

```python
def Ln_taylor(x, n):
    ##taylor series expansion
    
    # n is the number of terms
    # ln(1+x) = x + x^2/2 + ... + x^n/n
    sum = 0
            ##previous sum 
    for i in range(1,n):
        ##like in the slide
        sum = sum + ((-1)**(i+1))*(x**(i))/i
    #end for
    return sum
#end function
```
