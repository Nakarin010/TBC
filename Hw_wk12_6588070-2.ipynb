```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

```python
def getMinDP(A,n):
    #A is a list
    #n is the length of list
    if(n == 1):
        return A[0]
    elif(n == 2):
        if(A[0]<A[1]):
            return A[0]
        else:
            return A[1]
        #end if
    else:
        mid = n/2
        A_left = A[:,mid].copy()
        A_right = A[mid,:].copy()
        min_left = getMinDP(A_left,len(A_left))
        min_right = getMinDP(A_right,len(A_right))
        if(min_left<min_right):
            return min_left
        else:
            return min_right
```

```python
#for function and this for min and max

def opt_para(f,x1,x2,x3):
    a1 = (x2-x1)*(x2-x1)*(f(x2)-f(x3))
    a2 = (x2-x3)*(x2-x3)*(f(x2)-f(x1))
    b1 = (x2-x1)*(f(x2)-f(x3)) 
    b2 = (x2-x3)*(f(x2)-f(x1))
    c = (a1-a2)/(b1-b2)
    x4 = x2 - (0.5*c)
    return x4
```

```python
def para_max(f,x1,x2,x3,eps):
    xout_prev = x2
    x4 = opt_para(f,x1,x2,x3)
    print('x1 = %4.4f\tx2 = %4.4f\tx3 = %4.4f\tx4 = %4.4f' %(x1, x2, x3, x4))
    print('f(x1) = %4.4f\tf(x2) = %4.4f\tf(x3) = %4.4fItf(x4) = %4.4f' %(f(x1),f(x2),f(x3),f(x4)))
    if(x4>x2):
        if(f(x4) < f(x2)):
            xmax = x4
            x1 = x2
            x2 = x4
        else: #f(x4) > f(x2)
            xmax = x2
            x3 = x4
    else: #x4<x2
        if(f(x4) >f(x2)):
            xmax = x4
            x3 = x2
            x2 = x4
        else: #(f(x4) <f(x2)):
            xmax = x2
            x1 = x4
    xopt = xmax
    epa = np.abs((xopt - xout_prev)/xopt)*100
    print('xopt = %4.4f\tepa = %4.4f' %(xopt,epa))
    if(epa < eps):
        return xopt
    else:
        return para_max(x1,x2,x3,eps)
```
