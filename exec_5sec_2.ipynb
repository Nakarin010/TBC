2.  Given $ y= x^2 + 2x +5 $ plot graph of y(x), y'(x) and solve for maximum value of y(x). 

```python
from matplotlib import pyplot as plt
import numpy as np
from sympy import symbols, evalf, solve, diff

x,y = symbols('x y')
y = x**2 + 2*x + 5 


##calculate the first and second derivatives
dy_dx = diff(y,x)
dy_dx2 = diff(dy_dx,x)

## solve for maximum value of y(x)
x_max = solve(dy_dx, x)[0]
y_max = y.subs(x, x_max)

print(f"The maximum value of y is {y_max:.2f} when x is {x_max:.2f}")


#to make sure it's float and it can be use in linspace
xmnu = float(x_max.evalf())

##define range of x for plotting
x_values = np.linspace(xmnu - 2, xmnu + 2, 400) #center the range around the maximum
y_values = np.array([y.subs(x, val).evalf() for val in x_values])
y_prime_values = np.array([dy_dx.subs(x, val).evalf() for val in x_values])


# Plot y(x)
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='y(x) = x^2 + 2x + 5', color='blue')

# Plot y'(x)
plt.plot(x_values, y_prime_values, label="y'(x)", color='red')

#plot dot 
plt.plot(xmnu, y_max, marker='*', markersize=30, color='m',label=f'Maximum: ({x_max:.2f},{y_max:.2f})',zorder=5)
# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.grid(True)
plt.title('Plot of y(x) and y\'(x)')
plt.show()
```
